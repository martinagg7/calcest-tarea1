## **Parte 1: Paseo Aleatorio Simple en tiempo discreto**  

Un **paseo aleatorio simple** es un proceso estocástico definido por:

$$
S_t = S_{t-1} + X_t, \quad t = 1, 2, 3, \dots
$$

donde:
- $S_t$  representa la posición en el instante  $t$.
- $X_t$  es una variable aleatoria i.i.d. tales que:
  $$
  X_t =
  \begin{cases}

  -1, & \text{con probabilidad } \frac{1}{2} 
  \end{cases}
  $$

3. Representar gráficamente las trayectorias de las 200 simulaciones en la misma figura.

## Código parte 1

```python
#Parámetros
num_realizaciones = 200 
num_pasos = 100  
np.random.seed(987)
```

```python
#Matriz para almacenar todas las realizaciones
paseos_discretos = np.zeros((num_realizaciones, num_pasos + 1))

#Generar los paseos aleatorios
for i in range(num_realizaciones):
    pasos = np.random.choice([-1, 1], size=num_pasos)  
    paseos_discretos[i, 1:] = np.cumsum(pasos)  #Sumar pasos acumulados a partir de S0=0
```

```python
#Grafica con las 200 realizaciones
plt.figure(figsize=(10, 6))
for i in range(num_realizaciones):
    plt.plot(range(num_pasos + 1), paseos_discretos[i], alpha=0.3)


plt.xlabel("Tiempo (t)")
plt.ylabel("St")
plt.title("Paseo Aleatorio en Tiempo Discreto-200 realizaciones")
plt.grid(True)
plt.show()
```

## **Parte 2: Paseo Aleatorio Simple en tiempo continuo**  

Consideremos ahora la versión del **paseo aleatorio simple en tiempo continuo** donde $X_t \sim N(0,1)$ son variables aleatorias i.i.d. normales estándar.

```python
paseos_continuos = np.zeros((num_realizaciones, num_pasos + 1))
for i in range(num_realizaciones):
    pasos = np.random.normal(0, 1, size=num_pasos)   #N(0,1)
    paseos_continuos[i, 1:] = np.cumsum(pasos) 
```

```python
#Gráfica
plt.figure(figsize=(10, 6))
for i in range(num_realizaciones):
    plt.plot(range(num_pasos + 1), paseos_continuos[i], alpha=0.3,linewidth=0.7)

plt.xlabel("Tiempo (t)")
plt.ylabel("St")
plt.title("Paseo Aleatorio en Tiempo Continuo-200 realizaciones")
plt.grid(True)
plt.show()
```

**Explicación Teórica del Paseo Aleatorio en Tiempo Discreto**

*Media Esperada de  $S_t$*

Para calcular la esperanza de $S_t$, utilizamos la **propiedad lineal de la esperanza**:

$
E[S_t] = E[S_0] + \sum_{i=1}^{t} E[X_i]
$

Dado que \( S_0 = 0 \) y que la esperanza de \( X_t \) es:

$
E[X_t] = (+1) \cdot \frac{1}{2} + (-1) \cdot \frac{1}{2} = 0
$

se obtiene:

$
E[S_t] = 0 + \sum_{i=1}^{t} 0 = 0
$

Por lo tanto, **la media del paseo aleatorio en cualquier instante \( t \) es 0**.





*Varianza de $S_t$*

Dado que  $S_t$  es la suma de  t  variables  $X_t$ i.i.d, su varianza se obtiene como:

$
\text{Var}(S_t) = t \cdot \sigma_X^2
$

Como $X_t$ sigue una distribución de Bernoulli simétrica con valores \( \pm1 \) y probabilidad \( \frac{1}{2} \), su varianza es:

$
\sigma_X^2 = 1
$

Por lo tanto, se cumple que:

$
\text{Var}(S_t) = t
$


#### **Parte 4: Simulaciones de Montecarlo para obtener propiedades estadísticas**  

Repetir 5000 veces la simulación de la Parte 1 (es decir, se generarán 200 realizaciones del paseo aleatorio simple discreto en cada repetición). Para cada experimento:

1. Se generarán 200 realizaciones del paseo aleatorio simple discreto.
2. Se calcularán las medias muestrales y varianzas muestrales en $t = 5$, $t = 15$ y $t = 25$.
3. Se almacenarán los valores de media y varianza obtenidos en cada una de las 5000 repeticiones.

Al finalizar, se calcularán:

1. La media de las 5000 medias muestrales para cada $ t $.
2. La media de las 5000 varianzas muestrales para cada $ t $.

```python
#Definición de los parámetros y matrices para resultados
num_simulaciones = 5000  
num_realizaciones = 200   
num_pasos = 100           
t_interes = [5, 15, 25]


medias_montecarlo = np.zeros((num_simulaciones, len(t_interes)))
varianzas_montecarlo = np.zeros((num_simulaciones, len(t_interes)))


np.random.seed(987)
```

```python
#Generar 200 trayectorias y extraer los valores de t que nos interesan
for i in range(num_simulaciones):

    paseos_simulados = np.zeros((num_realizaciones, num_pasos + 1))
    
    for j in range(num_realizaciones):
        pasos = np.random.choice([-1, 1], size=num_pasos)  
        paseos_simulados[j, 1:] = np.cumsum(pasos) 
    
    
    S_t_select = paseos_simulados[:, t_interes]

    # media y varianza muestral en cada tiempo de interés
    medias_montecarlo[i, :] = np.mean(S_t_select, axis=0)
    varianzas_montecarlo[i, :] = np.var(S_t_select, axis=0, ddof=1)  # ddof=1 para varianza muestral
```

```python
#Calcular la media de las 5000 medias y varianzas muestrales
media_promedio = np.mean(medias_montecarlo, axis=0)
varianza_promedio = np.mean(varianzas_montecarlo, axis=0)

# Resultados
df_montecarlo = pd.DataFrame({
    "Tiempo t": t_interes,
    "Media Promedio de E[S_t]": media_promedio,
    "Varianza Promedio de Var[S_t]": varianza_promedio
})


print(df_montecarlo)
```
